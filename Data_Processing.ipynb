```python
temp = pd.read_csv('data/temperature_history.csv')
load = pd.read_csv('data/Load_history.csv')
load['zone_id']=load.zone_id.astype("category")
for i in range(4,28):
    load.iloc[:,i]= load.iloc[:,i].str.replace(",","").astype(float)
```

# Datasets are now complete

## Find which temperature station has highest correlation with demand for given station

```python
#We dont want to study correlation and pick station using test set information
#filter out test data
newTemp=temp.drop(temp[(temp['year'] == 2008) & (temp['month'] >= 1)].index)

#track how many days are in training set
nrow=newTemp[newTemp['station_id']==1].shape[0] #1582 days

#We want a matrix with each column representation a station, and each row representing 1 hour of 1 day

#create temporary var that will be overwritten
tempsForStation=np.zeros((nrow*24,1))

#for each of the 11 stations
for stationID in range(1,12):
    #grab data only for that station. Already sorted by date
    currentTemps=newTemp[newTemp['station_id']==stationID]
    #flatten() starts with first row and appends second row to end, then third row, etc. Make it a column vec
    currentTemps=np.reshape(currentTemps.iloc[:,4:].values.flatten(),(nrow*24,1))
    if(stationID==1):
        tempsForStation=currentTemps #overwrite temporary variable for first station
    else:
        tempsForStation=np.hstack((tempsForStation,currentTemps)) #add column to matrix
        
```

```python
#We dont want to study correlation and pick station using test set information
#filter out test data
newLoad=load.drop(load[(load['year'] == 2008) & (load['month'] >= 1)].index)

#track how many days are in training set
nrow=newLoad[newLoad['zone_id']==1].shape[0] #1582 days

#create container that holds which station id has highest correlation magnitude for the zone
HighestCorrStationID=[]


df_for_model = pd.DataFrame()

#for each of the 20 zones
for ZoneID in range(1,21):
    
    #grab data only for that zone and flatten as we did for station data
    current=newLoad[newLoad['zone_id']==ZoneID]
    current=current.iloc[:,4:]
    currentLoad=np.reshape(current.values.flatten(),(nrow*24,1))
    
    #append the zone data column to the station temp data matrix
    ZoneAndTemps=np.hstack((tempsForStation,currentLoad))
    #create correlation matrix using the magnitudes of correlation
    ZoneAndTemps=pd.DataFrame(ZoneAndTemps)
    corr=ZoneAndTemps.corr().abs()
    #index of maximum of last column represents station id most correlated to zone's load
    #delete last row since correlation is 1 with itself
    corr=corr.iloc[:-1,:]
    #grab max of last row. Add 1 since id's start with 1 but index starts at 0
    column=corr.iloc[:,11]
    HighestCorrStationID.append(column.idxmax()+1)
    
    # Append to dataframe to load into SARIMAX models
    df_for_model["Load"+str(ZoneID)] = currentLoad[:,0]
    df_for_model["Temp"+str(ZoneID)] = tempsForStation[:,column.idxmax()]
```
