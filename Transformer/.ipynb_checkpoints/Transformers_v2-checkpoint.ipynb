```python
# WARNING: If error is "super(type, obj): obj must be an instance or subtype of type". Just restart kernel.
train_data, val_data = get_data(training_set, test_percentage = 0.10)
model = TransAm().to(device)

print("Training Data shape:", train_data.shape)
print("Val Data shape:", val_data.shape)

criterion = nn.MSELoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=0.005 )
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, 1.0, gamma=0.98)

epochs = 2 # The number of epochs

for epoch in range(1, epochs + 1):
    epoch_start_time = time.time()
    model = train(model, optimizer, criterion, epoch, scheduler, train_data)
    
    # Print information about val loss
    if(epoch % 10 == 0):
        val_loss = plot_and_loss(model, criterion, val_data, epoch)
        predict_future(model, val_data,200)
    else:
        val_loss = evaluate(model, criterion, val_data)
        
    print('-' * 89)
    print('| end of epoch {:3d} | time: {:5.2f}s | valid loss {:5.5f} | valid ppl {:8.2f}'.format(epoch, (time.time() - epoch_start_time),
                                     val_loss, math.exp(val_loss)))
    print('-' * 89)

    scheduler.step() 
```
